```python
from datetime import datetime
```

FUNCION PARA INICIAR SESION

```python
def iniciar_sesion(usuarios):
    """
    Función que maneja el inicio de sesión de usuarios.
    Argumentos:
        usuarios (list): Lista de diccionarios con la información de los usuarios
    Retornos:
        dict: Usuario autenticado o None si no se pudo autenticar
    """
    while True:
        # Solicitar credenciales
        usuario = input("Ingrese nombre de usuario: ")
        contrasena = input("Ingrese contraseña: ")

        # Buscar usuario en la lista
        for user in usuarios:
            if user["nombre"] == usuario and user["contrasena"] == contrasena:
                print(f"Bienvenido {user['nombre']}")
                return user

        # Si no se encontró el usuario o la contraseña es incorrecta
        print("Credenciales inválidas. Intente nuevamente.")
```

FUNCION PARA REGISTRAR USUARIOS  

```python
def registrar_usuario(usuario_actual, usuarios):
    """
    Función que maneja el registro de nuevos usuarios.
    Solo los administradores pueden registrar nuevos usuarios.

    Argumentos:
        usuario_actual (dict): Usuario que intenta realizar el registro
        usuarios (list): Lista de diccionarios con la información de los usuarios

    Retorna:
        bool: True si el registro fue exitoso, False si no
    """
    # Verificar si es administrador
    if usuario_actual["rol"] != "admin":
        print("No tiene permisos para realizar esta operación")
        return False

    try:
        # Solicitar datos del nuevo usuario
        nuevo_nombre = input("Ingrese nombre del nuevo usuario: ")
        nueva_contrasena = input("Ingrese contraseña: ")
        nuevo_rol = input("Ingrese rol (admin/vendedor): ").lower()

        # Validar rol
        if nuevo_rol not in ["admin", "vendedor"]:
            print("Rol inválido. Debe ser 'admin' o 'vendedor'")
            return False

        # Generar nuevo ID
        nuevo_id = max(usuario["id"] for usuario in usuarios) + 1

        # Crear nuevo usuario
        nuevo_usuario = {
            "id": nuevo_id,
            "nombre": nuevo_nombre,
            "contrasena": nueva_contrasena,
            "rol": nuevo_rol
        }

        # Agregar a la lista de usuarios
        usuarios.append(nuevo_usuario)

        print("Usuario registrado exitosamente")
        return True

    except Exception as e:
        print(f"Error al registrar usuario: {str(e)}")
        return False
```

FUNCION PARA AGREGAR PRODUCTOS

```python
def agregar_producto(usuario_actual, productos):
    """
    Función que maneja el agregado o actualización de productos.
    Solo los administradores pueden agregar o actualizar productos.

    Argumentos:
        usuario_actual (dict): Usuario que intenta realizar la operación
        productos (list): Lista de diccionarios con la información de los productos

    Retorna:
        bool: True si la operación fue exitosa, False si no
    """
    # Verificar si es administrador
    if usuario_actual["rol"] != "admin":
        print("No tiene permisos para realizar esta operación")
        return False

    try:
        # Solicitar nombre del producto
        nombre = input("Ingrese nombre del producto: ")

        # Buscar si el producto ya existe
        producto_existente = None
        for producto in productos:
            if producto["nombre"].lower() == nombre.lower():
                producto_existente = producto
                break

        # Si el producto existe, actualizar cantidad y opcionalmente precios
        if producto_existente:
            print(f"\nProducto encontrado:")
            print(f"Cantidad actual: {producto_existente['cantidad']}")

            # Actualizar cantidad
            cantidad_adicional = int(input("Ingrese cantidad adicional: "))
            producto_existente['cantidad'] += cantidad_adicional

            # Preguntar si desea actualizar precios
            actualizar_precios = input("¿Desea actualizar los precios? (S/N): ").upper()
            if actualizar_precios == "S":
                producto_existente['precio_compra'] = float(input("Ingrese nuevo precio de compra: "))
                producto_existente['precio_venta'] = float(input("Ingrese nuevo precio de venta: "))

            print("\nInventario actualizado exitosamente")
            print(f"Nueva cantidad disponible: {producto_existente['cantidad']}")

        # Si el producto no existe, crear uno nuevo
        else:
            try:
                # Solicitar datos del nuevo producto
                precio_compra = float(input("Ingrese precio de compra: "))
                precio_venta = float(input("Ingrese precio de venta: "))
                cantidad = int(input("Ingrese cantidad inicial: "))
                tipo_unidad = input("Ingrese tipo de unidad: ")

                # Generar nuevo ID
                nuevo_id = max(producto["id"] for producto in productos) + 1 if productos else 1

                # Crear nuevo producto
                nuevo_producto = {
                    "id": nuevo_id,
                    "nombre": nombre,
                    "precio_compra": precio_compra,
                    "precio_venta": precio_venta,
                    "cantidad": cantidad,
                    "tipo_unidad": tipo_unidad
                }

                # Agregar a la lista de productos
                productos.append(nuevo_producto)
                print("\nProducto nuevo agregado exitosamente")

            except ValueError:
                print("Error: Los precios y cantidades deben ser números")
                return False

        return True

    except Exception as e:
        print(f"Error al procesar el producto: {str(e)}")
        return False
```

FUNCION PARA CONSULTAR PRODUCTOS

```python
def consultar_producto(productos):
    """
    Función que permite consultar productos del inventario,
    ya sea por código específico o listar todos.

    Argumentos:
        productos (list): Lista de diccionarios con la información de los productos

    Retorna:
        None
    """
    try:
        # Mostrar opciones de consulta
        print("\n=== CONSULTA DE PRODUCTOS ===")
        print("1. Buscar por código")
        print("2. Listar todos los productos")

        opcion = input("Seleccione una opción: ")

        # Buscar por código
        if opcion == "1":
            codigo = int(input("Ingrese código del producto: "))
            producto_encontrado = False

            for producto in productos:
                if producto["id"] == codigo:
                    print("\n=== DATOS DEL PRODUCTO ===")
                    print(f"ID: {producto['id']}")
                    print(f"Nombre: {producto['nombre']}")
                    print(f"Precio venta: ${producto['precio_venta']:,.2f}")
                    print(f"Cantidad disponible: {producto['cantidad']}")
                    print(f"Unidad: {producto['tipo_unidad']}")
                    producto_encontrado = True
                    break

            if not producto_encontrado:
                print("Producto no encontrado")

        # Listar todos los productos
        elif opcion == "2":
            if not productos:
                print("No hay productos en el inventario")
                return

            print("\n=== LISTADO DE PRODUCTOS ===")
            for producto in productos:
                print("\n-------------------")
                print(f"ID: {producto['id']}")
                print(f"Nombre: {producto['nombre']}")
                print(f"Precio venta: ${producto['precio_venta']:,.2f}")
                print(f"Cantidad disponible: {producto['cantidad']}")
                print(f"Unidad: {producto['tipo_unidad']}")

        else:
            print("Opción inválida")

    except ValueError:
        print("Error: El código debe ser un número")
    except Exception as e:
        print(f"Error al consultar productos: {str(e)}")
```

FUNCION PARA REALIZAR UNA VENTA

```python
def realizar_venta(usuario_actual, productos, ventas):
    """
    Función que maneja la realización de ventas.
    Solo los vendedores pueden realizar ventas.

    Argumentos:
        usuario_actual (dict): Usuario que intenta realizar la venta
        productos (list): Lista de diccionarios con la información de los productos
        ventas (list): Lista de diccionarios con el registro de ventas

    Retorna:
        bool: True si la venta fue exitosa, False si no
    """
    # Verificar si es vendedor
    if usuario_actual["rol"] != "vendedor":
        print("No tiene permisos para realizar esta operación")
        return False

    try:
        # Solicitar código de producto
        codigo_producto = int(input("Ingrese código del producto: "))

        # Buscar el índice del producto en el inventario
        indice_producto = -1
        for i in range(len(productos)):
            if productos[i]["id"] == codigo_producto:
                indice_producto = i
                break

        # Verificar si se encontró el producto
        if indice_producto == -1:
            print("Error: Producto no encontrado en el inventario")
            return False

        # Mostrar información del producto
        print(f"\nProducto: {productos[indice_producto]['nombre']}")
        print(f"Precio unitario: ${productos[indice_producto]['precio_venta']:,.2f}")
        print(f"Stock disponible: {productos[indice_producto]['cantidad']}")

        # Solicitar cantidad a vender
        cantidad_venta = int(input("Ingrese cantidad a vender: "))

        # Validar cantidad
        if cantidad_venta <= 0:
            print("Error: La cantidad debe ser mayor a 0")
            return False

        if cantidad_venta > productos[indice_producto]['cantidad']:
            print("Error: Stock insuficiente")
            print(f"Cantidad solicitada: {cantidad_venta}")
            print(f"Cantidad disponible: {productos[indice_producto]['cantidad']}")
            return False

        # Calcular total de la venta
        total_venta = cantidad_venta * productos[indice_producto]['precio_venta']

        # Actualizar inventario
        productos[indice_producto]['cantidad'] -= cantidad_venta

        # Registrar la venta
        nueva_venta = {
            "id": len(ventas) + 1,
            "fecha": datetime.now().strftime("%Y-%m-%d"),
            "vendedor_id": usuario_actual["id"],
            "producto_id": productos[indice_producto]["id"],
            "cantidad": cantidad_venta,
            "total": total_venta
        }

        ventas.append(nueva_venta)

        # Mostrar resumen de la venta
        print("\n=== Venta Realizada ===")
        print(f"Producto: {productos[indice_producto]['nombre']}")
        print(f"Cantidad vendida: {cantidad_venta}")
        print(f"Total: ${total_venta:,.2f}")
        print(f"Nueva cantidad en inventario: {productos[indice_producto]['cantidad']}")

        return True

    except ValueError:
        print("Error: Los valores ingresados deben ser números")
        return False
    except Exception as e:
        print(f"Error al procesar la venta: {str(e)}")
        return False
```

FUNCION PARA RESUMEN DE VENTAS

```python
def ver_resumen_ventas(usuario_actual, ventas, productos):
    """
    Función que muestra el resumen de todas las ventas realizadas.
    Solo los administradores pueden ver el resumen.

    Argumentos:
        usuario_actual (dict): Usuario que intenta ver el resumen
        ventas (list): Lista de diccionarios con la información de las ventas
        productos (list): Lista de diccionarios con la información de los productos

    Retorna:
        bool: True si la operación fue exitosa, False si no
    """
    # Verificar si es administrador
    if usuario_actual["rol"] != "admin":
        print("No tiene permisos para realizar esta operación")
        return False

    try:
        # Verificar si hay ventas
        if not ventas:
            print("No hay ventas registradas en el sistema")
            return False

        # Inicializar total general
        total_general = 0

        print("\n=== RESUMEN DE VENTAS ===")

        # Recorrer todas las ventas
        for venta in ventas:
            # Buscar información del producto vendido
            producto_vendido = None
            for producto in productos:
                if producto["id"] == venta["producto_id"]:
                    producto_vendido = producto
                    break

            if producto_vendido:
                print("\n-------------------")
                print(f"ID Venta: {venta['id']}")
                print(f"Fecha: {venta['fecha']}")
                print(f"Producto: {producto_vendido['nombre']}")
                print(f"Cantidad: {venta['cantidad']} {producto_vendido['tipo_unidad']}")
                print(f"Total: ${venta['total']:,.2f}")

                total_general += venta['total']

        # Mostrar total general
        print("\n====================")
        print(f"Total General: ${total_general:,.2f}")

        return True

    except Exception as e:
        print(f"Error al generar resumen de ventas: {str(e)}")
        return False
```
